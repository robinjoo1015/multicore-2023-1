```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
static_block_result = np.loadtxt('./static_block_result.csv', delimiter=',', skiprows=1, usecols=range(1,11))
```

```python
static_block_result.shape
```

```python
static_cyclic_result = np.loadtxt('./static_cyclic_result.csv', delimiter=',', skiprows=1, usecols=range(1,11))
```

```python
static_cyclic_result.shape
```

```python
dynamic_result = np.loadtxt('./dynamic_result.csv', delimiter=',', skiprows=1, usecols=range(1,11))
```

```python
dynamic_result.shape
```

```python
static_block_result
```

```python
np.mean(static_block_result, axis=0)
```

```python
np.mean(static_cyclic_result, axis=0)
```

```python
np.mean(dynamic_result, axis=0)
```

```python
parameters = {'axes.labelsize': 25,
              'axes.titlesize': 25,
              'xtick.labelsize': 25,
              'ytick.labelsize': 25,
              'ytick.direction':'in',
              'xtick.direction':'in',
             'legend.fontsize':25}
plt.rcParams.update(parameters)
```

```python
thread_num_list = [1,2,4,6,8,10,12,14,16,32]
```

```python
plt.figure(figsize=(20,12))
plt.errorbar(thread_num_list, np.mean(static_block_result, axis=0), np.std(static_block_result, axis=0), fmt='-o', capsize=10, markersize=10, label='static (block)', c='r')
plt.errorbar(thread_num_list, np.mean(static_cyclic_result, axis=0), np.std(static_cyclic_result, axis=0), fmt='-o', capsize=10, markersize=10, label='static (cyclic)', c='g')
plt.errorbar(thread_num_list, np.mean(dynamic_result, axis=0), np.std(dynamic_result, axis=0), fmt='-o', capsize=10, markersize=10, label='dynamic', c='blue')
plt.xticks(thread_num_list)
plt.legend()
plt.title('Execution Time')
plt.xlim(0,33)
plt.xlabel('$\#\ of\ threads$')
plt.ylabel('$ms$')
plt.show()
```

```python
static_block_perf = 1. / static_block_result
static_cyclic_perf = 1. / static_cyclic_result
dynamic_perf = 1. / dynamic_result
```

```python
np.mean(static_block_perf, axis=0)
```

```python
np.mean(static_cyclic_perf, axis=0)
```

```python
np.mean(dynamic_perf, axis=0)
```

```python
plt.figure(figsize=(20,12))
plt.errorbar(thread_num_list, np.mean(static_block_perf, axis=0), np.std(static_block_perf, axis=0), fmt='-o', capsize=10, markersize=10, label='static (block)', c='r')
plt.errorbar(thread_num_list, np.mean(static_cyclic_perf, axis=0), np.std(static_cyclic_perf, axis=0), fmt='-o', capsize=10, markersize=10, label='static (cyclic)', c='g')
plt.errorbar(thread_num_list, np.mean(dynamic_perf, axis=0), np.std(dynamic_perf, axis=0), fmt='-o', capsize=10, markersize=10, label='dynamic', c='blue')
plt.xticks(thread_num_list)
plt.legend()
plt.title('Performance')
plt.xlim(0,33)
plt.xlabel('$\#\ of\ threads$')
plt.ylabel('$1/ms$')
plt.show()
```
