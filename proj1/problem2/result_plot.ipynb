```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
mat500_result = np.loadtxt('./mat500_result.csv', delimiter=',', skiprows=1, usecols=range(1,11))
```

```python
mat500_result.shape
```

```python
mat500_result
```

```python
np.mean(mat500_result, axis=0)
```

```python
parameters = {'axes.labelsize': 25,
              'axes.titlesize': 25,
              'xtick.labelsize': 25,
              'ytick.labelsize': 25,
              'ytick.direction':'in',
              'xtick.direction':'in',
             'legend.fontsize':25}
plt.rcParams.update(parameters)
```

```python
thread_num_list = [1,2,4,6,8,10,12,14,16,32]
```

```python
plt.figure(figsize=(20,12))
plt.errorbar(thread_num_list, np.mean(mat500_result, axis=0), np.std(mat500_result, axis=0), fmt='-o', capsize=10, markersize=10, label='mat500')
plt.xticks(thread_num_list)
plt.legend()
plt.title('Execution Time')
plt.xlim(0,33)
plt.xlabel('$\#\ of\ threads$')
plt.ylabel('$ms$')
plt.show()
```

```python
mat500_perf = 1. / mat500_result
```

```python
np.mean(mat500_perf, axis=0)
```

```python
plt.figure(figsize=(20,12))
plt.errorbar(thread_num_list, np.mean(mat500_perf, axis=0), np.std(mat500_perf, axis=0), fmt='-o', capsize=10, markersize=10, label='mat500')
plt.xticks(thread_num_list)
plt.legend()
plt.title('Performance')
plt.xlim(0,33)
plt.xlabel('$\#\ of\ threads$')
plt.ylabel('$1/ms$')
plt.show()
```
